```python
lower_bound = 0.0
upper_bound = 4194304.0
total_numbers = 10**9  # One billion
batch_size = 10**6  # One million
batch_no=1

# Function to generate random numbers in batches
def generate_random_numbers(n):
    for _ in range(0, n, batch_size):
        yield [random.randint(lower_bound, upper_bound) for _ in range(min(batch_size, n - _))]

# Generate one billion random numbers in batches
all_random_numbers = []
for batch in generate_random_numbers(total_numbers):
    print("Batch number :",batch_no)
    all_random_numbers.extend(batch)
    batch_no = batch_no+1
```

```python
file_name = "data/uniform_data_large.txt"

# Open the file in write mode and save the list elements separated by a new line
with open(file_name, 'w') as file:
    for number in all_random_numbers:
        file.write(str(number) + '\n')

print("List of random numbers saved to", file_name)
```

```python
def scatterplot(df,col1,col2s,colors,point_size,isLine,yline,makersize=40,is_ylimt=False,ylimit=100000):
    plt.figure(figsize=(18, 9))

    for i, cols in enumerate(col2s):
        plt.scatter(df[col1], df[cols],s=point_size,color=colors[i],label=cols)

    # Set labels for the axes

        
    plt.xlabel(col1)
    plt.ylabel(' '.join(col2s))

    # Set a title for the plot
    plt.title("Scatter Plot of "+col1 +" vs. "+' '.join(col2s)+ "(ns) for Read Only")
    if is_ylimt:
        plt.ylim(0,ylimit)
    
    if(isLine):
        plt.axhline(y=yline, color='red', linestyle='--', label='Threshold')


    plt.legend(markerscale= makersize,loc='upper left')

    plt.savefig("NFL Scatterplot read only small.png")
```

```python
def scatterplot_insert(df,df_insert,col1,col2s,colors,point_size,isLine,yline,makersize=40,is_ylimt=False,ylimit=100000):
    plt.figure(figsize=(18, 9))

    for i, cols in enumerate(col2s):
        plt.scatter(df[col1], df[cols],s=point_size,color=colors[i],label=cols)

    # Set labels for the axes
    plt.scatter(df_insert[col1], [5000 for _ in range(len(df_insert))],s=point_size, color='red', label='Inserted Data')


        
    plt.xlabel(col1)
    plt.ylabel(' '.join(col2s))

    # Set a title for the plot
    plt.title("Scatter Plot of "+col1 +" vs. "+' '.join(col2s)+ "(ns) for Insert and Read")
    if is_ylimt:
        plt.ylim(0,ylimit)
    
    if(isLine):
        plt.axhline(y=yline, color='red', linestyle='--', label='Threshold')


    plt.legend(markerscale= makersize,loc='upper left')

    plt.savefig("NFL Scatterplot insert read.png")
```

```python
def linechart(df,col1,col2s,colors,point_size,isLine,yline,makersize=40,is_ylimt=False,ylimit=100000):
    plt.figure(figsize=(18, 9))

    for i, cols in enumerate(col2s):
        plt.plot(df[col1], df[cols],color=colors[i],label=cols)

    # Set labels for the axes

        
    plt.xlabel(col1)
    plt.ylabel(' '.join(col2s))

    # Set a title for the plot
    plt.title("Scatter Plot of "+col1 +" vs. "+' '.join(col2s)+ "(ns) for Read Only")
    if is_ylimt:
        plt.ylim(0,ylimit)
    
    if(isLine):
        plt.axhline(y=yline, color='red', linestyle='--', label='Threshold')


    plt.legend(markerscale= makersize,loc='upper left')

    plt.savefig("NFL Line plot read only 25.png")
```
